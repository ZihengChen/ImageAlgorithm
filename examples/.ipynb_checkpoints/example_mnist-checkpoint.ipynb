```python
from ImageAlgoKD import *
```

## 1. Load 1000 MNIST Data Points

```python
# read 1000 mnist points
points = Points(np.genfromtxt("../data/mnist.csv",delimiter=','))

# plot first 10 mnist points
plt.figure(figsize=(10,1),facecolor="w")
for i in range(10):
    plt.subplot(1,10,i+1)
    plt.imshow(points.cords[i].reshape(28,28).T)
    plt.axis('off')
```

## 2. Define Clustering Parameters

```python
# define clustering parameters
ia = ImageAlgoKD(MAXDISTANCE        = 2.0,
                 KERNEL_R           = 0.7,
                 KERNEL_R_NORM      = 0.7,
                 KERNEL_R_POWER     = 2.0,
                 DECISION_RHO_KAPPA = 4.7,
                 DECISION_NHD       = 0.7,
                 CONTINUITY_NHD     = 0.7)
ia.setInputsPoints(points)
```

## 3. Run and Plot Results

```python
ia.run("numpy",1,100)

# plot decision
plt.figure(figsize=(6,4),facecolor="w")

x,y,slt = ia.points.rho, ia.points.nhd, ia.points.isSeed
plt.plot(x,y, ".")
plt.scatter( x[slt], y[slt], marker="s", s=100, facecolors='none', edgecolors="k")

plt.axvline(ia.DECISION_RHO,color='k',linestyle="--",linewidth=1)
plt.axhline(ia.DECISION_NHD,color='k',linestyle="--",linewidth=1)
plt.xlabel("Density")
plt.ylabel("Seperation")
plt.title("Decistion Plot")
plt.savefig("../plots/mnist_decision.png",dpi=300)
```

```python
# plot clusters
nDemo = 10
nClust = ia.points.clusterID.max()+1

plt.figure(figsize=(nDemo,nClust),facecolor="w")

for icl in range(nClust):
    sl = ia.points.clusterID==icl
    x = ia.points.cords[sl]
    for i in range(nDemo):
        plt.subplot(nClust,nDemo,nDemo*icl+i+1)
        plt.imshow(x[i].reshape(28,28).T)
        plt.axis('off')
        
plt.savefig("../plots/mnist.png",dpi=300)
```

## 4. Check Opencl and Numpy Gives Consistent Result

```python
ia.run("opencl",0,1)
gpu = ia.points.clusterID

ia.run("numpy",1,32)
cpu = ia.points.clusterID

identical = (gpu==cpu)
identical.all()
```
